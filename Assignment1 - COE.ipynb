```python
uber_ride = uber_ride.rename(columns = {'START_DATE*' : 'START_DATE', 'END_DATE*' : 'END_DATE', 'CATEGORY*' : 'CATEGORY', 'START*' : 'START'})
uber_ride
```

```python
#Boolean Indexing and Filtering
a = uber_ride[uber_ride['CATEGORY'] == "Business"]
a
```

```python
uber_ride.sort_values(by='MILES*', ascending=False).head(5)
```

```python
# Replace all missing PURPOSE values with "Not Specified"
uber_ride['PURPOSE*'] = uber_ride['PURPOSE*'].fillna('Not Specified')
print(uber_ride['PURPOSE*'].isnull().sum())
```

```python
# Converting  START_DATE* and END_DATE* to datetime format
uber_ride['START_DATE']=pd.to_datetime(uber_ride['START_DATE'],format="%m/%d/%Y %H:%M", errors='coerce')
uber_ride['END_DATE']=pd.to_datetime(uber_ride['END_DATE'],format="%m/%d/%Y %H:%M", errors='coerce')
uber_ride.head()
# Create a new column TRIP_DURATION (in hours)
uber_ride['TRIP_DURATION']=(uber_ride['END_DATE']-uber_ride['START_DATE']).dt.total_seconds()/3600
uber_ride[['START_DATE','END_DATE','TRIP_DURATION']].head()
```

```python
uber_ride.sort_values(by='MILES*', ascending=False).head(5)
```

```python
# Sort dataset by distance in descending order
sorted=uber_ride.sort_values(by='MILES*',ascending=False)
# Display first few rows
sorted.head()
```
