```python
# Calculate total and average sales per month
monthly_sales=sales_data[['Jan','Feb','Mar','Apr']]
# Total sales per month
total_sales = monthly_sales.sum()
print(total_sales)
```

```python
# Average sales per month
average_sales = monthly_sales.mean()
print( average_sales)
```

```python
# Total yearly sales for all stores combined
total_yearly_sales_all_stores = sales_data[['Jan','Feb','Mar','Apr']].sum()
print(total_yearly_sales_all_stores)
```

```python
# Add Total_Sales column (sum of Jan to Apr)
sales_data['TOTAL_SALES']=sales_data[['Jan','Feb','Mar','Apr']].sum(axis=1)
sales_data.head()
```

```python
# Top 5 stores with highest Total_Sales
top_5= sales_data.sort_values(by='TOTAL_SALES',ascending=False).head(5)
top_5[['store_id','TOTAL_SALES']]
```

```python
# Bottom 5 stores with lowest Total_Sales
bottom_5=sales_data.sort_values(by='TOTAL_SALES',ascending=True).head(5)
bottom_5[['store_id','TOTAL_SALES']]
```

```python
# Find store with maximum sales in March
max_march=sales_data.loc[sales_data['Mar'].idxmax(), 'store_id']
max_march
```

```python
# Find store with maximum sales in April
min_apr=sales_data.loc[sales_data['Apr'].idxmin(), 'store_id']
min_apr
```

```python
# Filter stores from California that had sales > 20 in any month.
sales_data[(sales_data['city']=='California')&((sales_data['Jan']>20)|(sales_data['Feb']>20)|(sales_data['Mar']>20)|(sales_data['Apr']>20))][['store_id','Jan','Feb','Mar','Apr','TOTAL_SALES']]
```
